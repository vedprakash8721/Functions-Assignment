 Qno 1. What is the difference between a function and a method in Python?
-  
Function:
A function is a standalone block of code designed to perform a specific task. It is independent of any object and can be called directly by its name with required arguments.

 Method:
A method is a function that is associated with an object (usually an instance of a class). It implicitly receives the object itself (via self) and typically operates on or modifies the data within that object.


````python
# Function (standalone)
def greet(name):
    return f"Hello, {name}!"

# Method (inside a class)
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # method
        return f"Hello, {self.name}!"
````

#Qno 4, What is the purpose of the `return` statement in a Python function?
-  The purpose of the return statement in a Python function is to send a result or value back to the part of the program that called the function. When Python encounters return, it immediately exits the function and passes the specified value back to the caller.

 Without a return statement, a function will return None by default, meaning it doesn’t produce any usable output.

 Why is return important?

#Qno 5. What are iterators in Python and how do they differ from iterables?
- What is an Iterable?
An iterable is any Python object capable of returning its elements one at a time, allowing you to loop over it. Common examples include lists, tuples, strings, dictionaries, and sets. Iterables implement the special method __iter__() which returns an iterator.

 What is an Iterator?
An iterator is an object that represents a stream of data; it produces the next value each time you call its __next__() method. Iterators keep track of their current position internally and raise a StopIteration exception when there are no more elements to return.

 Key Difference
Iterable: Has the ability to return an iterator (__iter__() method). It’s like a container that can provide access to its elements sequentially.

 Iterator: Is the object that actually performs the traversal over the iterable’s elements, producing one item at a time using __next__().



````python
my_list = [1, 2, 3]     # Iterable

my_iter = iter(my_list)  # Iterator obtained from iterable

print(next(my_iter))     # Outputs: 1
print(next(my_iter))     # Outputs: 2
print(next(my_iter))     # Outputs: 3
# next(my_iter) now raises StopIteration
````

#Qno 6. Explain the concept of generators in Python and how they are defined.
- Generators are a special kind of iterator in Python that produce values one at a time, on demand, instead of computing and storing the entire sequence at once. This makes them highly memory-efficient, especially for large datasets or infinite sequences.

 Generators are created using generator functions, which look like regular functions but use the yield keyword to return a value and pause their state. When the generator is resumed, it continues execution right after the last yield. This allows lazy evaluation — values are generated only when requested.

 How to define a generator function:

````python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
````

When you call count_up_to(5), it returns a generator object. Each time you iterate over it, the function runs until it hits yield and returns the current value, then pauses until the next request.

Why use generators?
Memory efficiency: No need to store large sequences in memory.

Improved performance: Values are computed only as needed.

Simpler code: Easier to write iterators without managing state manually.

#Qno 7.  What are the advantages of using generators over regular functions?
-  Memory Efficiency:
Generators yield values one at a time, so they don’t store the entire result in memory. Regular functions typically return complete data structures, which can consume a lot of memory for large outputs.

  Lazy Evaluation:
Generators compute values only when requested, avoiding unnecessary calculations upfront. Regular functions compute everything before returning.

 Better Performance with Large Data:
Because generators produce items on demand, programs can start processing results immediately without waiting for the full computation.

 Simplified Code for Iteration:
Generators maintain their state automatically between yields, making it easier to write iterable logic without manually managing loops or indices.

 Support for Infinite Sequences:
Generators can represent infinite data streams, which is impossible with regular functions returning finite collections.

#Qno 9.  Explain the purpose and usage of the `map()` function in Python.
- Purpose of map()
The map() function is designed to apply a specified function to each item of one or more iterables (like lists or tuples) and return a new iterator with the results. It enables concise and efficient transformation of data without writing explicit loops.
function: A function that takes as many arguments as there are iterables.

 iterable: One or more sequences to which the function is applied element-wise.

 The result of map() is an iterator, so you typically convert it into a list or iterate over it directly.

````python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
````

#Qno 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- 1. map()
Purpose: Applies a given function to each element of one or more iterables, returning an iterator with the transformed elements.

 Use case: When you want to transform or modify every item in a collection.

 Example: Square each number in a list.

````python
numbers = [1, 2, 3]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9]
````

 - 2. filter()
Purpose: Applies a function that returns True or False to each element and returns an iterator with only the elements where the function returns True.

   Use case: When you want to select or filter items from a collection based on a condition.

  Example: Select even numbers from a list.

````python
numbers = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4]
````

#Qno 1. . Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.


````python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
numbers = [10, 15, 22, 33, 40]
print(sum_even_numbers(numbers))
````

#Qno 4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example: Check primes from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num)
````

#Qno 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

````python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)
````

#Qno 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
````

#Qno 7.  Implement a generator function that reads a file line by line and yields each line as a string.

````python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
````

#Qno 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

````python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)  # Output: [32.0, 68.0, 98.6, 212.0]
````

#Qno 10. . Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, s))

# Example usage
input_str = "Hello, World!"
result = remove_vowels(input_str)
print(result)  # Output: Hll, Wrld!
````
